### ¿Qué vamos a aprender?

Antes de aplicar cualquier modelo de forecasting, necesitamos responder una pregunta fundamental:

> **¿La serie de tiempo es estacionaria?**

Una serie es **estacionaria** cuando su **media**, **varianza** y **autocorrelación** son constantes en el tiempo.  
La mayoría de los modelos clásicos (ARIMA, Holt-Winters) **lo requieren**. Si no lo verificamos, nuestros pronósticos serán basura.

## 1. 📦 Librerías y carga de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.ticker import FuncFormatter
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf
from scipy.stats import gaussian_kde

# Estilo global de gráficas
plt.rcParams.update({
    'figure.facecolor': '#F8FAFC',
    'axes.facecolor':   '#F8FAFC',
    'axes.grid':        True,
    'grid.color':       '#E2E8F0',
    'grid.linewidth':   0.8,
    'font.size':        11,
})

BLUE   = '#2563EB'
RED    = '#DC2626'
GREEN  = '#16A34A'
PURPLE = '#7C3AED'

money_fmt = FuncFormatter(lambda x, _: f'${x/1e6:.1f}M')

print('✅ Librerías cargadas correctamente')
```

```python
from pathlib import Path

# Detecta la ruta del CSV sin importar desde dónde se lanzó Jupyter
# Busca 'output/sales_history.csv' subiendo niveles hasta encontrarlo
def find_csv(filename='sales_history.csv', max_levels=4):
    base = Path().resolve()
    for _ in range(max_levels):
        candidate = base / 'output' / filename
        if candidate.exists():
            return candidate
        base = base.parent
    raise FileNotFoundError(
        f"No se encontró '{filename}' en output/. "
        "Asegúrate de haber corrido main.py primero."
    )

csv_path = find_csv()
print(f'✅ CSV encontrado en: {csv_path}')

# Carga sin parsear fechas para inspeccionar columnas reales
df_raw = pd.read_csv(csv_path, nrows=2)
print(f'Columnas encontradas: {df_raw.columns.tolist()}')

# Detecta automáticamente la columna de fecha
date_col = next(
    (c for c in df_raw.columns if 'date' in c.strip().lower()),
    None
)
if date_col is None:
    raise ValueError(f"No se encontró columna de fecha. Columnas disponibles: {df_raw.columns.tolist()}")
print(f'Columna de fecha detectada: "{date_col}"')

# Carga completa con la columna correcta
df = pd.read_csv(csv_path, parse_dates=[date_col])
df.rename(columns={date_col: 'date'}, inplace=True)  # normaliza el nombre
df.sort_values('date', inplace=True)

print(f'\nFilas       : {len(df):,}')
print(f'Período     : {df["date"].min().date()} → {df["date"].max().date()}')
print(f'SKUs únicos : {df["sku_id"].nunique()}')
print(f'Canales     : {df["channel"].unique().tolist()}')
df.head(3)
```

## 2. 🔎 Construcción de la Serie de Tiempo

Para este análisis vamos a trabajar con los **ingresos semanales totales** — la suma de todos los SKUs y canales.  
Esta es la serie que típicamente analizaría un analista antes de modelar la demanda agregada de la empresa.

```python
# Agregamos por semana: suma de ingresos de todos los SKUs y canales
serie = (
    df.groupby('date')['revenue']
    .sum()
    .resample('W-MON')      # frecuencia semanal
    .sum()
)
serie.name = 'revenue'

print(f'Observaciones : {len(serie)}')
print(f'Media semanal : ${serie.mean():,.0f}')
print(f'Mín           : ${serie.min():,.0f}')
print(f'Máx           : ${serie.max():,.0f}')
serie.head()
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(serie.index, serie.values, color=BLUE, linewidth=1.2, alpha=0.85)
ax.yaxis.set_major_formatter(money_fmt)
ax.set_title('Ingresos Semanales Totales — Serie Original', fontsize=14, fontweight='bold')
ax.set_xlabel('Fecha')
ax.set_ylabel('Ingresos')
plt.tight_layout()
plt.show()

# ¿Qué ves? La serie sube año con año → hay tendencia.
# También hay picos repetidos (Buen Fin, Navidad) → hay estacionalidad.
# Vamos a confirmarlo con herramientas formales.
```

## 3. 👁️ Prueba Visual: Media y Varianza Móviles

La forma más intuitiva de detectar no-estacionariedad es graficar la **media móvil** y la **desviación estándar móvil**:

| Comportamiento | Interpretación |
|---|---|
| Media móvil **sube o baja** | ❌ La media no es constante → **tendencia** |
| Media móvil **plana** | ✅ Posiblemente estacionaria |
| Desv. estándar **crece** | ❌ Varianza no constante → **heterocedasticidad** |
| Desv. estándar **plana** | ✅ Varianza constante |

```python
WINDOW = 12  # 12 semanas ≈ 3 meses

roll_mean = serie.rolling(WINDOW).mean()
roll_std  = serie.rolling(WINDOW).std()

fig = plt.figure(figsize=(14, 8))
gs  = gridspec.GridSpec(2, 1, hspace=0.45)

# Panel 1: Serie + media móvil
ax1 = fig.add_subplot(gs[0])
ax1.plot(serie.index, serie.values, color=BLUE, alpha=0.5, linewidth=1, label='Serie original')
ax1.plot(roll_mean.index, roll_mean.values, color=RED, linewidth=2.5,
         label=f'Media móvil ({WINDOW} semanas)')
ax1.yaxis.set_major_formatter(money_fmt)
ax1.set_title('¿La media es constante en el tiempo?', fontsize=12, fontweight='bold')
ax1.legend(loc='upper left')

# Panel 2: Desv. estándar móvil
ax2 = fig.add_subplot(gs[1])
ax2.plot(roll_std.index, roll_std.values, color=GREEN, linewidth=2,
         label=f'Desv. estándar móvil ({WINDOW} semanas)')
ax2.yaxis.set_major_formatter(money_fmt)
ax2.set_title('¿La varianza es constante en el tiempo?', fontsize=12, fontweight='bold')
ax2.legend(loc='upper left')

fig.suptitle('Prueba Visual de Estacionariedad', fontsize=15, fontweight='bold')
plt.show()

# 💡 Observa que la media móvil tiene pendiente positiva → tendencia creciente
# La desv. estándar también crece → varianza no constante
```

## 4. 🔬 Pruebas Estadísticas: ADF y KPSS

La prueba visual es útil pero subjetiva. Las pruebas estadísticas nos dan un **veredicto objetivo**.

Usaremos **dos pruebas** con hipótesis opuestas — esto elimina la ambigüedad:

| Prueba | Hipótesis nula (H₀) | p-valor pequeño significa... |
|---|---|---|
| **ADF** | La serie **NO** es estacionaria | ✅ **SÍ** es estacionaria |
| **KPSS** | La serie **SÍ** es estacionaria | ❌ **NO** es estacionaria |

### ⚠️ Detalle crítico: el parámetro `regression`

Ambas pruebas tienen un parámetro que define **qué estructura determinista** se asume en la serie.  
Si no los alineamos, estamos comparando mundos distintos — como medir en °C y °F sin convertir.

| `regression` | ADF | KPSS | Significa |
|---|---|---|---|
| `'c'` | ✅ | ✅ | Solo constante (serie estacionaria alrededor de una media) |
| `'ct'` | ✅ | ✅ | Constante + tendencia lineal |
| `'nc'` | ✅ | — | Sin constante ni tendencia |

> **Regla:** Si tu serie tiene tendencia visible → usa `regression='ct'` en **ambas** pruebas.  
> Nuestra serie claramente sube año a año → usamos `'ct'`.

**Veredicto combinado:**

| ADF (`ct`) | KPSS (`ct`) | Conclusión |
|---|---|---|
| Rechaza H₀ | No rechaza H₀ | 🟢 Estacionaria en tendencia |
| No rechaza H₀ | Rechaza H₀ | 🔴 No estacionaria — diferenciación necesaria |
| Ambas rechazan | — | 🟡 Ambiguo — revisar |
| Ninguna rechaza | — | 🟡 Ambiguo — revisar |

```python
import warnings
from statsmodels.tools.sm_exceptions import InterpolationWarning

def run_adf(s, alpha=0.05, regression='c'):
    stat, p, lags_used, _, crit, _ = adfuller(s.dropna(), autolag='AIC', regression=regression)
    reg_label = {'c': 'solo constante', 'ct': 'constante + tendencia', 'nc': 'sin constante'}
    print('─' * 56)
    print(f'  ADF | regression="{regression}" ({reg_label.get(regression, "")})')
    print(f'  H₀: La serie NO es estacionaria | Lags (AIC): {lags_used}')
    print('─' * 56)
    print(f'  Estadístico : {stat:.4f}')
    print(f'  p-valor     : {p:.4f}')
    print(f'  Críticos    : { {k: round(v,4) for k,v in crit.items()} }')
    if p < alpha:
        print(f'  ✅ p < {alpha} → Se rechaza H₀ → Estacionaria (bajo este modelo)')
    else:
        print(f'  ❌ p ≥ {alpha} → No se rechaza H₀ → NO estacionaria')
    return p < alpha

def run_kpss(s, alpha=0.05, regression='c'):
    reg_label = {'c': 'solo constante', 'ct': 'constante + tendencia'}
    with warnings.catch_warnings(record=True) as caught:
        warnings.simplefilter('always')
        stat, p, _, crit = kpss(s.dropna(), regression=regression, nlags='auto')
    interp = [w for w in caught if issubclass(w.category, InterpolationWarning)]
    print('─' * 56)
    print(f'  KPSS | regression="{regression}" ({reg_label.get(regression, "")})')
    print(f'  H₀: La serie SÍ es estacionaria')
    print('─' * 56)
    print(f'  Estadístico : {stat:.4f}')
    print(f'  p-valor     : {p:.4f}')
    print(f'  Críticos    : { {k: round(v,4) for k,v in crit.items()} }')
    if interp:
        print(f'  ⚠️  InterpolationWarning: estadístico fuera de la tabla.')
        print(f'     p-valor es un límite, no un valor exacto. Interpreta con cautela.')
    if p < alpha:
        print(f'  ❌ p < {alpha} → Se rechaza H₀ → NO estacionaria')
    else:
        print(f'  ✅ p ≥ {alpha} → No se rechaza H₀ → Estacionaria (bajo este modelo)')
    return p >= alpha

def veredicto(adf_r, kpss_r, regression):
    print(f'\n{"═"*56}')
    print(f'  VEREDICTO — regression="{regression}"')
    print(f'{"═"*56}')
    if adf_r and kpss_r:
        print('  🟢 Ambas pruebas: ESTACIONARIA (bajo este modelo)')
    elif not adf_r and not kpss_r:
        print('  🔴 Ambas pruebas: NO ESTACIONARIA → diferenciación necesaria')
    else:
        print('  🟡 Resultado mixto → apóyate en la prueba visual')
    print(f'{"═"*56}\n')
```

```python
# ── RONDA 1: regression='c' → ¿Es estacionaria EN NIVELES? ──────────────────
# Esta es la pregunta honesta: sin permitirle tendencia al modelo.
print('\n🔬 RONDA 1 — regression="c" (sin tendencia en el modelo)\n')
adf_c  = run_adf(serie,  regression='c')
print()
kpss_c = run_kpss(serie, regression='c')
veredicto(adf_c, kpss_c, 'c')
```

```python
# ── RONDA 2: regression='ct' → ¿Es estacionaria ALREDEDOR DE UNA TENDENCIA? ─
# Pregunta distinta: le permitimos tendencia y vemos si lo restante es ruido.
print('\n🔬 RONDA 2 — regression="ct" (con tendencia en el modelo)\n')
adf_ct  = run_adf(serie,  regression='ct')
print()
kpss_ct = run_kpss(serie, regression='ct')
veredicto(adf_ct, kpss_ct, 'ct')
```

```python
# ── Interpretación conjunta ──────────────────────────────────────────────────
print('╔══════════════════════════════════════════════════════╗')
print('║         INTERPRETACIÓN: LAS DOS RONDAS              ║')
print('╠══════════════════════════════════════════════════════╣')
print('║  regression="c"  → NO estacionaria   ← VERDAD      ║')
print('║  regression="ct" → estacionaria       ← TRAMPA      ║')
print('╠══════════════════════════════════════════════════════╣')
print('║  Con "ct" el test le PERMITE tener tendencia        ║')
print('║  y luego pregunta si lo que queda es estacionario.  ║')
print('║  No responde si la serie es estacionaria AS-IS.     ║')
print('╠══════════════════════════════════════════════════════╣')
print('║  La evidencia visual es concluyente:                ║')
print('║  media móvil creciente + varianza creciente         ║')
print('║  = NO ESTACIONARIA EN NIVELES.                      ║')
print('╚══════════════════════════════════════════════════════╝')
print()
print('💡 Regla de oro:')
print('   Aunque ADF/KPSS con "ct" sugieran estacionariedad,')
print('   si la rolling mean tiene pendiente visible,')
print('   la serie NO es estacionaria en niveles.')
print('   La prueba visual gana cuando el modelo del test')
print('   ya incorpora la no-estacionariedad que buscas detectar.')
```

## 5. 💊 Tratamiento: Diferenciación

La **diferenciación** es la técnica más común para eliminar tendencia y lograr estacionariedad.

$$\Delta y_t = y_t - y_{t-1}$$

En lugar de modelar el valor absoluto de ventas, modelamos el **cambio** entre períodos.  
Esto es exactamente el parámetro `d` en ARIMA(`p`, **`d`**, `q`).

```python
diff1 = serie.diff().dropna()
diff2 = serie.diff().diff().dropna()

fig, axes = plt.subplots(3, 1, figsize=(14, 11))

configs = [
    (serie, 'Serie Original',             BLUE),
    (diff1, 'Primera Diferencia (d=1)',   PURPLE),
    (diff2, 'Segunda Diferencia (d=2)',   RED),
]

for ax, (s, title, color) in zip(axes, configs):
    ax.plot(s.index, s.values, color=color, linewidth=1.2)
    ax.axhline(s.mean(), color='black', linestyle='--',
               linewidth=1.2, alpha=0.7, label=f'Media: ${s.mean():,.0f}')
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.yaxis.set_major_formatter(money_fmt)
    ax.legend(loc='upper right', fontsize=9)

fig.suptitle('Efecto del Diferenciado sobre la Tendencia', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Verificamos que la primera diferencia ya es estacionaria
print('Pruebas sobre la PRIMERA DIFERENCIA:\n')
adf_d1   = run_adf(diff1)
print()
kpss_d1  = run_kpss(diff1)
```

## 6. 📊 ACF: La Firma Visual de la Estacionariedad

La **Función de Autocorrelación (ACF)** mide qué tan correlacionada está la serie consigo misma en distintos rezagos.

| Patrón en la ACF | Significado |
|---|---|
| Decaimiento **muy lento** | ❌ Serie no estacionaria — hay tendencia |
| Decaimiento **rápido** (primeros 2-3 lags) | ✅ Serie estacionaria |
| Picos cada **n períodos** | 🔄 Hay estacionalidad de período n |

```python
LAGS = 40

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

plot_acf(serie,  lags=LAGS, ax=ax1, color=BLUE,   alpha=0.05)
plot_acf(diff1,  lags=LAGS, ax=ax2, color=PURPLE,  alpha=0.05)

ax1.set_title('ACF — Serie Original\n→ Decaimiento lento = tendencia', 
              fontsize=12, fontweight='bold')
ax2.set_title('ACF — Primera Diferencia\n→ Decaimiento rápido = más estacionaria', 
              fontsize=12, fontweight='bold')

for ax in (ax1, ax2):
    ax.set_xlabel('Rezago (semanas)')
    ax.set_ylabel('Autocorrelación')

fig.suptitle('ACF: Antes y Después de Diferenciar', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 💡 En el próximo video veremos también la PACF para identificar
#    los parámetros p y q de un modelo ARIMA.
```
